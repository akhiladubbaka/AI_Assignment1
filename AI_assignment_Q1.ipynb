```python
from collections import deque
def is_goal(state):
    return state == ['L', 'L', 'L', '_', 'R', 'R', 'R']
def get_possible_moves(state):
    moves = []
    for i in range(len(state)):
        if state[i] == 'R':
            if i + 1 < len(state) and state[i + 1] == '_':
                new_state = state[:]
                new_state[i], new_state[i + 1] = new_state[i + 1], new_state[i]
                moves.append(new_state)
            if i + 2 < len(state) and state[i + 2] == '_' and state[i + 1] in ['L']:
                new_state = state[:]
                new_state[i], new_state[i + 2] = new_state[i + 2], new_state[i]
                moves.append(new_state)
        elif state[i] == 'L':
            if i - 1 >= 0 and state[i - 1] == '_':
                new_state = state[:]
                new_state[i], new_state[i - 1] = new_state[i - 1], new_state[i]
                moves.append(new_state)
            if i - 2 >= 0 and state[i - 2] == '_' and state[i - 1] in ['R']:
                new_state = state[:]
                new_state[i], new_state[i - 2] = new_state[i - 2], new_state[i]
                moves.append(new_state)
    return moves
def bfs(start_state):
    visited = set()
    queue = deque()
    queue.append((start_state, [start_state]))
    visited.add(tuple(start_state))

    while queue:
        current, path = queue.popleft()
        if is_goal(current):
            return path
        for move in get_possible_moves(current):
            move_tuple = tuple(move)
            if move_tuple not in visited:
                visited.add(move_tuple)
                queue.append((move, path + [move]))
    return None
def dfs(start_state):
    visited = set()
    stack = [(start_state, [start_state])]
    visited.add(tuple(start_state))

    while stack:
        current, path = stack.pop()
        if is_goal(current):
            return path
        for move in get_possible_moves(current):
            move_tuple = tuple(move)
            if move_tuple not in visited:
                visited.add(move_tuple)
                stack.append((move, path + [move]))
    return None
start = ['R', 'R', 'R', '_', 'L', 'L', 'L']
bfs_result = bfs(start)
dfs_result = dfs(start)
print("\n--- BFS Solution ---")
if bfs_result:
    for step in bfs_result:
        print(' '.join(step))
else:
    print("No BFS solution found.")

print("\n--- DFS Solution ---")
if dfs_result:
    for step in dfs_result:
        print(' '.join(step))
else:
    print("No DFS solution found.")
```
